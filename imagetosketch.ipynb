**Importing module and file**



```python
import cv2
from google.colab import files

```

```python
pic = cv2.imread("luffy.png")
```

**below is the original.**

```python
cv2_imshow(pic)
```

**converting to grayscale**

```python
graypic = cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY)
cv2_imshow(graypic)
```

**Inverting grayscale image**

```python
invertedpic = 255 - graypic
cv2_imshow(invertedpic)
```

**Bluring the image**

```python
blurpic = cv2.GaussianBlur(invertedpic, (21, 21), 0)
cv2_imshow(blurpic)
```

**Finally converting to pencil sketch**

```python
invertblurpic = 255 - blurpic
pencilsketch = cv2.divide(graypic, invertblurpic, scale=256.0)
cv2_imshow(pencilsketch)
```

**the original and the sketch.**

```python
cv2_imshow(pic)
cv2_imshow(pencilsketch)
```
